```python
#handle dumbp json file
json_text = [json.loads(line) for line in open('Commercial_10.json', 'r')]
json_text
```

```python
#frame the data
parse_json_text=[]
for i in json_text:
 parse_json_text.append(i)
df=pd.DataFrame(parse_json_text).head(5)
df
```

## Takes a column of lists and removes the list and creates more rows using pandas.explode

```python
mod_df = df.loc[:,['address','city','brokers']]
for i in mod_df.columns:
    new_df = mod_df.explode(i)
mod_df.head(5)
```

## Deal with the dictionary left behind with json.normalize

```python
flat_dct=pd.json_normalize(json.loads(new_df.to_json(orient="records")))
flat_dct
```

## Output

```python
output = flat_dct
```

```python
output.to_excel("commercial_info.xlsx", index=False)
```
