# Bring demographic data in front
Organize the more basic demographic info to the front to enhance the Clarity and Readability of the spreadsheet afterward

```python
for k in {1:5}: 
 for i in df_modified.iloc[:, 13:17]:
  col = df_modified.pop(i) # Remove the column i from the DataFrame and store it in 'col'
  df_modified.insert(k, col.name, col) # Insert the 'col' back into the DataFrame at position k

df_modified
```

# Unpivot for desired format of the dataset

```python
id_vars = df_modified.columns[:4] # Identify columns to be used as identifier variables and value variables for pandas.df.melt
value_vars =df_modified.columns[5:]
df_melted = df_modified.melt(id_vars= id_vars, value_vars=value_vars, var_name="Question + Subquestion", value_name="Answers") #unpivot, from wide to long
df_melted
```

# Find the number of responses for each question 

```python
df_melted['Question'] = df_melted['Question + Subquestion'].apply(str).apply(lambda x: x.split('-')[0]) # Extract the 'Question' from 'Question + Subquestion' for further use
df_gr = df_melted[df_melted['Answers'].notna()] # Filter out the non-answered elements, those will create noise
df_gr
```

```python
df_gr = df_gr.groupby("Question")["Index"].nunique().reset_index() #Groups the DataFrame df_gr by the "Question" column then counts the number of unique values in the "Index" column for each group, which give the number of respondent for a particular question.
df_gr.rename(columns={"Index":"Respondents for the question"}, inplace=True)
df_gr
```

# Merge this new dataframe with previous df_melted

```python
df_merged = pd.merge(left = df_melted, right = df_gr, how ="left", left_on= "Question", right_on="Question")
print("Original", len(df_melted))
print("Merged Response count", len(df_merged))
df_merged
```

# Find the number of the same response for each question

```python
df_gr_2 = df_melted[df_melted['Answers'].notna()] # Filter out the non-answered elements, those will create noise in next steps
df_gr_2 = df_gr_2.groupby(["Question + Subquestion","Answers"])["Index"].nunique().reset_index() # group by question and count distinct respondent ID
df_gr_2.rename(columns={"Index":"Same Answer"}, inplace=True)
df_gr_2
```

# Merge this new dataframe with previous "number of respondents" dataframe

```python
df_merged_2 = pd.merge(left = df_merged, right = df_gr_2, how ="left", left_on= ["Question + Subquestion","Answers"], right_on=["Question + Subquestion","Answers"])
df_merged_2["Same Answer"].fillna(0, inplace=True)
#reconciliate the original dataframe length and final  dataframe length
print("Original", len(df_melted)) 
print("Merged Answer count", len(df_merged_2))
df_merged_2
```

# Finalize format

```python
df_merged_final = df_merged_2.copy()
# Get the column names
columns = df_merged_final.columns.tolist()
# Swap the positions of "Answers" and "Question" for redability
columns[6], columns[5] = columns[5], columns[6]
# Update the DataFrame with the new column order
df_merged_final = df_merged_final[columns]
df_merged_final = df_merged_final[df_merged_final['Answers'].notna()]
df_merged_final = df_merged_final.drop(columns='Question + Subquestion') # Delete this column (this is basically Question + Answer)
df_merged_final
```

# Export
